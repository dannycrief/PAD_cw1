```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('DATA/PAD_03_PD.csv', delimiter=';', index_col='ID')
```

```python
# Zadanie 1
df['Country'].value_counts()
```

```python
# Zadanie 2
df['owned_goods'] = df['owns_car']+df['owns_TV']+df['owns_house']+df['owns_Phone']
df
```

```python
# Zadanie 3
men_mean = round(df[df['gender'] == "M"]['owned_goods'].mean(), 2)
women_mean = round(df[df['gender'] == "K"]['owned_goods'].mean(), 2)

print(f"Średnia posiadanych dóbr pomiędzy mężczyznami: {men_mean}")
print(f"Średnia posiadanych dóbr pomiędzy kobietami: {women_mean}")
```

```python
# Zadanie 4
df2=df.dropna(how='all')
df_zad4 = df2.groupby(['Country'])[['owned_goods']].mean()
df_zad4['min_age'] = df2.groupby(['Country'])[['Age']].min()
df_zad4
```
